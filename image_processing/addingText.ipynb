```python
from PIL import Image, ImageDraw, ImageFont
```

```python
plankton = Image.open('plankton.jpeg')
```

```python
drawable = ImageDraw.Draw(plankton)
```

```python
myfont = ImageFont.truetype('font.ttf', 20)
```

```python
drawable.text((50, 50), "plankton!! The Devil", (255, 0, 0), font=myfont)
```
